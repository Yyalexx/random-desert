```python
import pandas as pd
import json
```

```python
recipe_df = pd.read_csv('./recipe_df.csv')
```

```python
# Список id рецептов для экспорта в JSON
recipe_id_list = list(range(16))
#recipe_id_list = [0,1,2,3,4]
```

```python
dict_json = {} # словарь словарей рецептов для экспорта в JSON
dict_json['recipe_id']={}
for recipe_id in recipe_id_list:
    # текущий рецепт из списка для экспорта
    recipe = recipe_df.loc[recipe_df.recipe_id==recipe_id]
    recipe_dict={}
    ingr_dict={}
    # формирование словаря ингредиентов текущего рецепта
    for i in range(len(recipe)):
        ingr_dict[recipe.ingr_name.iloc[i]] = recipe.ingr_qwn.iloc[i]
    # формирование словаря текущего рецепта
    recipe_dict["cuisine"]=recipe.cuisine.iloc[0]
    recipe_dict["meal_mode"]=recipe.meal_mode.iloc[0]
    recipe_dict["recipe_name"]=recipe.recipe_name.iloc[0]
    recipe_dict["portions"]=str(recipe.portions.iloc[0])
    recipe_dict["calories"]=str(recipe.calories.iloc[0])
    recipe_dict["proteinContent"]=str(recipe.proteinContent.iloc[0])
    recipe_dict["fatContent"]=str(recipe.fatContent.iloc[0])
    recipe_dict["carbohydrateContent"]=str(recipe.carbohydrateContent.iloc[0])
    recipe_dict["ingredients"]=ingr_dict
    recipe_dict["text"]=recipe.text.iloc[0]
    # добавление словаря текущего рецепта 
    # в список словарей рецептов для экспорта в JSON
    dict_json['recipe_id'][str(recipe_id)]=recipe_dict
```

```python
# #экспорт JSON файла
with open('./recipe.json', 'w') as outfile:
    json.dump(dict_json, outfile, ensure_ascii=False)
```
